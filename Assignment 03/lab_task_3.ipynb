```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Part 1 — Data Generation
1. Generate 3,000 wait times using a uniform distribution between 2 and 12 minutes.
2. Generate 3,000 wait times using a normal distribution with:
o Mean = 8
o Std Dev = 2
3. Ensure no negative wait times in the normal distribution (replace negatives with 0).

```python
np.random.seed(100)

# uniform distribution between 2 and 12 minutes
regular_wait_time = np.random.uniform(2, 12, 3000)

# normal distribution with: o Mean = 8 and Std Dev = 2
busy_wait_time = np.random.normal(8, 2, 3000)

# Ensure no negative
busy_wait_time = np.maximum(busy_wait_time, 0)
```

Part 2 — Visualization
1. Plot histograms for both distributions.
2. Show them:
 In the same figure (two subplots), OR Overlaid with transparency.

```python
plt.figure(figsize=[12, 5])
# Regular days
plt.subplot(1, 2, 1)
plt.hist(regular_wait_time, bins=30, edgecolor='black', color='blue', alpha=0.5)
plt.title('Regular Days')
plt.xlabel('Wait Time')
plt.ylabel('Number of Customers')
plt.grid(True)

# Busy days
plt.subplot(1, 2, 2)
plt.hist(busy_wait_time, bins=30, edgecolor='black', color='orange', alpha=0.5)
plt.title('Busy Days ')
plt.xlabel('Wait Time')
plt.ylabel('Number of Customers')
plt.grid(True)


plt.show()
```

Part 3 — Statistical Analysis
For both datasets calculate:
 Mean
 Median
 Standard deviation
 Maximum wait time

```python

# Calculate for regular Day
reg_mean = np.mean(regular_wait_time)
reg_median = np.median(regular_wait_time)
reg_std = np.std(regular_wait_time)
reg_max = np.max(regular_wait_time)

# # Calculate for Busy Day
busy_mean = np.mean(busy_wait_time)
busy_median = np.median(busy_wait_time)
busy_std = np.std(busy_wait_time)
busy_max = np.max(busy_wait_time)

# Printing results

print("Regular Days ")
print("  Mean: ",reg_mean)
print("  Median: ",reg_median)
print("  Std Dev: ",reg_std)
print("  Max: ",reg_max)

print("Busy Days:")
print("  Mean: ",busy_mean)
print("  Median: ",busy_median)
print("  Std Dev: ",busy_std)
print("  Max: ",busy_max)
```
