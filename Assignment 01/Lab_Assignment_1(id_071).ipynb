```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Create two 3Ã—3 matrices with random integers in Python.
A = np.random.randint(-10, 10, (3, 3))
B = np.random.randint(-10, 10, (3, 3))

print(A)
print(B)
```

```python
# Perform addition, subtraction, and multiplication operations on these matrices.
add = A + B
sub = A - B
mul = A @ B

print(add)
print(sub)
print(mul)
```

```python
#Compute the determinant, inverse, and rank of one of the matrices.
#determinant
det_A = np.linalg.det(A)

# Inverse
if np.linalg.det(A) != 0:
    inv_A = np.linalg.inv(A)
else:
    inv_A = "Matrix is singular, cannot compute inverse"

# Rank of a matrix
rank_A = np.linalg.matrix_rank(A)

print("Determinant of A:", det_A)
print("Inverse of A:\n", inv_A)
print("Rank of A:", rank_A)
```

```python
#Generate and Plot Two Random Vectors:
#a. Create two vectors, each containing 15 random floats of your choice.
#b. Plot these vectors on the same graph using matplotlib to compare how
# Label the axes appropriately and give the plot a title.

v1 = np.random.uniform(-1, 1, 15)
v2 = np.random.uniform(-1, 1, 15)

print(v1)
print(v2)


x = np.arange(len(v1))
print(x)

plt.plot(x, v1, marker='o', linestyle='-', label="Vector 1")
plt.plot(x, v2, marker='s', linestyle='--', label="Vector 2")


plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Comparison of Two Random Vectors")
plt.legend()
plt.grid(True)

plt.show()
```
