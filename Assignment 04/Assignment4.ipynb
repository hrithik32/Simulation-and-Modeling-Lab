Part 2- Simulation Logic
1. Write a function to check whether a point is inside the circle centered at the origin with radius R/10.
2. Use two different random methods:
random.uniform() for half of the points
numpy.random.rand()
3. Keeр a running count of:
Points inside the circle
(scaled to your range) for the other half
Points outside the circle

```python
def estimated_pi(num, r):
    inside = 0
    half = num // 2

    # First half: random.uniform
    for i in range(half):
        x = random.uniform(0, r)
        y = random.uniform(0, r)
        if (x**2 + y**2 <= r**2):
            inside += 1

    # Second half: numpy.random.rand
    for i in range(num - half):
        x = np.random.rand() * r
        y = np.random.rand() * r
        if (x**2 + y**2 <= r**2):
            inside += 1

    pi = 4 * inside / num
    return inside, pi
```
